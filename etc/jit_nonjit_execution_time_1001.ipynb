```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_without_jit(n):
    total = 0
    for i in range(n):
        total += (i % 7) * (i % 13)
    return total

def compute_with_jit(n):
    i = np.arange(n)
    return np.sum((i % 7) * (i % 13))
```

```python
sizes = np.array([10**i for i in range(4, 8)])

times_wo = []
times_w = []

for n in sizes:
    start = time.time()
    compute_without_jit(n)
    times_wo.append(time.time() - start)
    
    start = time.time()
    compute_with_jit(n)
    times_w.append(time.time() - start)

times_wo = np.array(times_wo)
times_w = np.array(times_w)
```

```python
plt.figure(figsize=(6, 5))

plt.plot(sizes, times_wo, marker='o', label="without JIT")
plt.plot(sizes, times_w, marker='o', label="with JIT")

plt.xscale('log')
plt.title("Execution Time vs Input Size")
plt.xlabel("Input size")
plt.ylabel("Time (seconds)")

plt.legend()
plt.grid(False)
plt.show()
```
