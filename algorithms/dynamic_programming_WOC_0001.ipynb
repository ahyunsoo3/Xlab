```python
import math
```

```python
def template_forward_1d(n):
    dp = [0] * (n+1)
    dp[0] = 1
    dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

print(template_forward_1d(10))
```

```python
def template_backward_1d(cost):
    n = len(cost)
    dp = [0] * (n+1)
    dp[n] = 0
    for i in range(n-1, -1, -1):
        dp[i] = cost[i] + min(
            dp[i+1],
            dp[i+2] if i+2 <= n else math.inf
        )
    return dp[0]

print(template_backward_1d([1,2,3,4]))
```

```python
def template_grid_dp(rows, cols):
    dp = [[0]*cols for _ in range(rows)]
    dp[0][0] = 1
    for i in range(rows):
        for j in range(cols):
            if i == 0 and j == 0:
                continue
            from_up = dp[i-1][j] if i > 0 else 0
            from_left = dp[i][j-1] if j > 0 else 0
            dp[i][j] = from_up + from_left
    return dp[rows-1][cols-1]

print(template_grid_dp(3, 3))
```

```python
def template_knapsack(weights, values, capacity):
    dp = [0] * (capacity+1)
    for i in range(len(weights)):
        weight = weights[i]
        value = values[i]
        for c in range(capacity, weight-1, -1):
            dp[c] = max(
                dp[c],
                value + dp[c-weight]
            )
    return dp[capacity]

print(template_knapsack([1,3,4], [10,20,30], 4))
```

```python
def template_string_dp(a, b):
    m, n = len(a), len(b)
    dp = [[0]*(n+1) for _ in range(m+1)]
    for i in range(1, m+1):
        for j in range(1, n+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(
                    dp[i-1][j],
                    dp[i][j-1]
                )
    return dp[m][n]

print(template_string_dp("abcde", "ace"))
```

```python
def template_memo(n):
    memo = {}
    def dp(state):
        if state in memo:
            return memo[state]
        if state <= 1:
            return state
        result = dp(state-1) + dp(state-2)
        memo[state] = result
        return result
    return dp(n)

print(template_memo(10))
```

```python
def template_space_optimized(n):
    prev = 0
    curr = 1
    for i in range(2, n+1):
        new = prev + curr
        prev = curr
        curr = new
    return curr

print(template_space_optimized(10))
```
