# Fibonacci Numbers

```python
def fib_memo(n, memo={}):
    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

print("Fib Memoization result:", fib_memo(10))
```

```python
def fib_tab(n):
    if n <= 1:
        return n

    dp = [0] * (n+1)
    dp[1] = 1

    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

print("Fib Tabulation result:", fib_tab(10))
```

# Climbing Stairs

You can climb 1 or 2 steps. How many ways to reach step n?

```python
def climb_stairs(n):
    if n <= 2:
        return n

    dp = [0]*(n+1)
    dp[1] = 1
    dp[2] = 2

    for i in range(3, n+1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

print("Ways to climb:", climb_stairs(5))
```

# Coin Change

Find minimum coins needed to make an amount.

Example: <br>
coins = [1,2,5], amount = 11 <br>
Answer = 3 (5+5+1)

```python
def coin_change(coins, amount):

    dp = [float('inf')] * (amount+1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount+1):
            dp[i] = min(dp[i], dp[i-coin] + 1)

    if dp[amount] == float('inf'):
        return -1

    return dp[amount]

print("Minimum coins:", coin_change([1,2,5], 11))
```

# Longest Increasing Subsequence

Find Longest Increasing Subsequence Length

Array: [10,9,2,5,3,7,101,18]
Answer: 4 → [2,3,7,101]

```python
def length_of_LIS(nums):

    n = len(nums)
    dp = [1] * n

    for i in range(n):
        for j in range(i):

            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)

print("LIS length:", length_of_LIS([10,9,2,5,3,7,101,18]))
```

# 0/1 Knapsack

weights = [1,3,4,5] <br>
values = [1,4,5,7] <br>
capacity = 7 <br>

Max value = 9

```python
def knapsack(weights, values, capacity):

    n = len(weights)

    dp = [[0]*(capacity+1) for _ in range(n+1)]

    for i in range(1, n+1):

        for w in range(capacity+1):

            if weights[i-1] <= w:

                dp[i][w] = max(
                    dp[i-1][w],
                    values[i-1] + dp[i-1][w-weights[i-1]]
                )

            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

print("Knapsack max value:", knapsack([1,3,4,5], [1,4,5,7], 7))
```

# Longest Common Subsequence 

text1 = "abcde" <br>
text2 = "ace" <br>

Answer: 3 → ace


```python
def lcs(text1, text2):

    m = len(text1)
    n = len(text2)

    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(1, m+1):

        for j in range(1, n+1):

            if text1[i-1] == text2[j-1]:

                dp[i][j] = dp[i-1][j-1] + 1

            else:

                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

print("LCS length:", lcs("abcde", "ace"))
```
