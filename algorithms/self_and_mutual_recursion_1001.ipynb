```python
print("===== 1. Recursion Example: Factorial =====")

def factorial(n):
    """Single recursion: function calls itself"""
    
    # Base case
    if n == 0 or n == 1:
        return 1
    
    # Recursive case
    return n * factorial(n - 1)


# Test factorial
for i in range(6):
    print(f"factorial({i}) = {factorial(i)}")
```

```python
print("\n===== 2. Mutual Recursion Example: Even/Odd =====")

def is_even(n):
    """Mutual recursion: calls is_odd"""
    
    if n == 0:
        return True
    
    return is_odd(n - 1)


def is_odd(n):
    """Mutual recursion: calls is_even"""
    
    if n == 0:
        return False
    
    return is_even(n - 1)


# Test even/odd
for i in range(6):
    print(f"{i} is even: {is_even(i)}")
```

```python
print("\n===== 3. Mutual Recursion with Call Trace =====")

def is_even_trace(n):
    print(f"is_even({n}) called")
    
    if n == 0:
        return True
    
    return is_odd_trace(n - 1)


def is_odd_trace(n):
    print(f"is_odd({n}) called")
    
    if n == 0:
        return False
    
    return is_even_trace(n - 1)


print("Trace result:", is_even_trace(4))
```

```python
print("\n===== 4. Recursion Example: Nested List Sum =====")

data = [1, [2, 3], [4, [5, 6]]]

def recursive_sum(lst):
    """Recursion used for nested structures"""
    
    total = 0
    
    for item in lst:
        if isinstance(item, list):
            total += recursive_sum(item)
        else:
            total += item
    
    return total


print("Nested sum =", recursive_sum(data))


print("\n===== DONE =====")
```
