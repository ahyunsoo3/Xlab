```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_edges_from([
    ("x > 0?", "y = 1"),
    ("x > 0?", "y = -1"),
    ("y = 1", "return y"),
    ("y = -1", "return y"),
])

nx.draw(G, with_labels=True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# Add edges representing control flow
G.add_edges_from([
    ("Start", "x > 0?"),               # Entry
    ("x > 0?", "x % 2 == 0?"),         # True branch of while
    ("x > 0?", "Return x"),            # False branch (exit loop)
    
    ("x % 2 == 0?", "x = x - 2"),      # If true
    ("x % 2 == 0?", "x = x - 1"),      # If false
    
    ("x = x - 2", "x > 0?"),           # Loop back
    ("x = x - 1", "x > 0?"),           # Loop back
])

plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# Add edges representing control flow
G.add_edges_from([
    ("Start", "x > 0?"),
    
    ("x > 0?", "Return positive"),      # True branch
    ("x > 0?", "x == 0?"),              # False branch
    
    ("x == 0?", "Return zero"),         # True branch
    ("x == 0?", "Return negative"),     # False branch
])

plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000)
plt.show()
```
